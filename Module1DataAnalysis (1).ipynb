Heatmap

```python
bool_df = df[["blueFirstBlood", "blueWins", "redFirstBlood"]]
sns.heatmap(bool_df.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation between Blue First Blood, Blue Wins, and Red First Blood")
plt.show()
```

Bar Plot with only Numeric variables

```python
df['blueWins'] = df['blueWins'].astype(bool)

numeric_df = df.select_dtypes(include="number")

correlations = numeric_df.corrwith(df['blueWins'])
correlations.sort_values().plot(
    kind="barh",
    figsize=(8,5)
)

plt.title("Correlation with blueWins")
plt.xlabel("Correlation")
plt.ylabel("Numeric Variables")

plt.show()
```

Making sure that Blue Wins is a Boolean

```python
from pandas.api.types import is_bool_dtype
df['blueWins'] = df['blueWins'].astype(bool)
is_bool_A = is_bool_dtype(df['blueWins'])
print(f"Is column A boolean? {is_bool_A}")
```
